```python

x=data.drop(['Class'],axis=1)
x.head()
```

```python
y=data['Class']
y.head()
```

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

K means

```python
from sklearn.neighbors import KNeighborsClassifier
neighbor=KNeighborsClassifier(n_neighbors=5)
```

```python

x_train_cleaned = x_train.dropna()
y_train_cleaned = y_train[x_train_cleaned.index]

neighbor.fit(x_train_cleaned, y_train_cleaned)
```

```python
predictions=neighbor.predict(x_test)
print(predictions)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test, predictions)
ac=accuracy_score(y_test, predictions)
print(cm)
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
```

```python
# Get the indices of rows where y_train is not NaN
not_nan_indices = y_train[~y_train.isnull()].index


x_train = x_train.loc[not_nan_indices]
y_train = y_train.loc[not_nan_indices]

tree.fit(x_train, y_train)
```

```python

predictions = tree.predict(x_test)

print("Predictions:", predictions)

from sklearn.tree import export_text
tree_rules = export_text(tree, feature_names=x.columns.tolist())
print("\nDecision Tree Rules:\n", tree_rules)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


accuracy = accuracy_score(y_test, predictions)
precision = precision_score(y_test, predictions)
recall = recall_score(y_test, predictions)
f1 = f1_score(y_test, predictions)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```
